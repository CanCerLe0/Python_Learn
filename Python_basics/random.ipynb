```python
#random是伪随机，它的随机数是可以通过种子被重现的
import random

x = random.getstate()
y = random.randint(1, 100)
z = random.randint(1, 100)
a = random.randint(1, 100)
print("第一次随机数：", y, z, a)
random.setstate(x)
y = random.randint(1, 100)
z = random.randint(1, 100)
a = random.randint(1, 100)
print("第二次随机数：", y, z, a) 
```

```python
#random模块的种子是可以被设置的，设置相同的种子会产生相同的随机数序列
#random模块的随机数生成器是基于梅森旋转算法的，这是一种高效的伪随机数生成算法，具有较长的周期和良好的统计性质。
random.seed(10)
x = random.randint(1, 100)
y = random.randint(1, 100)
z = random.randint(1, 100)
print("第一次随机数：", x, y, z)
random.seed(10)
x = random.randint(1, 100)
y = random.randint(1, 100)
z = random.randint(1, 100)
print("第二次随机数：", x, y, z)
```
