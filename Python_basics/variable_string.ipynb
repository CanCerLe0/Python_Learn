```python
x = 3
print(x)
y = 'Hello, World!'
print(y)
hi_123 = 'Hello, Python!' #变量名可以定义为字母、数字和下划线的组合，但必须以字母或下划线开头，不能以数字开头。
print(hi_123)
变量名 = '这是一个变量名' #变量名可以使用中文，但不建议使用，因为可能会导致代码的可读性和维护性降低。
print(变量名)
```

```python
print("D:\three\two\one") #路径中的反斜杠被python当成转义字符了，\t被当成了制表符，所以输出结果不对
```

```python
print("D:\\three\\two\\one") #用反斜杠来转译路径中的反斜杠
print(r"D:\three\two\one") #在字符串前面加上r，表示原始字符串，里面的转义字符不再起作用
```

```python
#反斜杠在末尾时表示续行符，表示下一行也是字符串的一部分
print("Life is short, \n\
let's learn python") 
```

```python
#Triple quted string 长字符串，表现为三引号（'''或"""）括起来的字符串，可以跨行，字符串中的换行符会被保留
print('''Life is short,
let's learn python''')
```

```python
print(520+1314) #数值的加法是数值的相加，所以结果是1834
print('520'+'1314') #字符串的加法是字符串的拼接，而不是数值的加法
print('520'*3) #字符串的乘法是字符串的重复，而不是数值的乘法
```
