```python
x = 3
print(x)
y = 'Hello, World!'
print(y)
hi_123 = 'Hello, Python!' #变量名可以定义为字母、数字和下划线的组合，但必须以字母或下划线开头，不能以数字开头。
print(hi_123)
变量名 = '这是一个变量名' #变量名可以使用中文，但不建议使用，因为可能会导致代码的可读性和维护性降低。
print(变量名)
```

```python
print(0b100)  # 二进制整数
print(0o100)  # 八进制整数
print(100)    # 十进制整数
print(0x100)  # 十六进制整数
```

```python
print(123.456)    # 数学写法
print(1.23456e2)  # 科学计数法
```

```python
print(0.1)
print(0.1 + 0.2)  # 浮点数运算可能会有精度问题
print(0.1 + 0.2 == 0.3)  # False
```

```python
a = 45        # 定义变量a，赋值45
b = 12        # 定义变量b，赋值12
print(a, b)   # 45 12
print(a + b)  # 57
print(a - b)  # 33
print(a * b)  # 540
print(a / b)  # 3.75
```

```python
a = 100
b = 123.45
c = 'hello, world'
d = True
print(type(a))  # <class 'int'>
print(type(b))  # <class 'float'>
print(type(c))  # <class 'str'>
print(type(d))  # <class 'bool'>
```

```python
a = 100
b = 123.45
c = '123'
d = '100'
e = '123.45'
f = 'hello, world'
g = True
print(float(a))         # int类型的100转成float，输出100.0
print(int(b))           # float类型的123.45转成int，输出123
print(int(c))           # str类型的'123'转成int，输出123
print(int(c, base=16))  # str类型的'123'按十六进制转成int，输出291
print(int(d, base=2))   # str类型的'100'按二进制转成int，输出4
print(float(e))         # str类型的'123.45'转成float，输出123.45
print(bool(f))          # str类型的'hello, world'转成bool，输出True
print(int(g))           # bool类型的True转成int，输出1
print(chr(a))           # int类型的100转成str，输出'd'
print(ord('d'))         # str类型的'd'转成int，输出100
```

```python
print("D:\three\two\one") #路径中的反斜杠被python当成转义字符了，\t被当成了制表符，所以输出结果不对
```

```python
print("D:\\three\\two\\one") #用反斜杠来转译路径中的反斜杠
print(r"D:\three\two\one") #在字符串前面加上r，表示原始字符串，里面的转义字符不再起作用
```

```python
#反斜杠在末尾时表示续行符，表示下一行也是字符串的一部分
print("Life is short, \n\
let's learn python") 
```

```python
#Triple quted string 长字符串，表现为三引号（'''或"""）括起来的字符串，可以跨行，字符串中的换行符会被保留
print('''Life is short,
let's learn python''')
```

```python
print(520+1314) #数值的加法是数值的相加，所以结果是1834
print('520'+'1314') #字符串的加法是字符串的拼接，而不是数值的加法
print('520'*3) #字符串的乘法是字符串的重复，而不是数值的乘法
```
